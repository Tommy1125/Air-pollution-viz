```python
def readfile(name):
    df = pd.read_excel('Global-Monthly-Sales-Data-External-2023-02-27.xlsx', sheet_name = name)
    df = df.iloc[2:, 1:81]
    df2 = pd.DataFrame()
    columns= df.iloc[2:7,0:1].transpose()
    for i in range(0,76,19):
        df3 = df.iloc[2:7,i+15:i+19].transpose()
        df3 = pd.concat([columns,df3])
        df3.columns = df3.iloc[0]
        df3 = df3[1:]
        df3['Country'] = name
        df3['Year']=df.iloc[1,0:5].unique()[1:5]
        df3['Month'] = df.iloc[0,i+15]
        df2 = pd.concat([df2, df3])
    return df2
```

```python
_df= pd.DataFrame()
df2= pd.DataFrame()
for name in names:
    df = readfile(name)
    df2 = pd.concat([df2,df])
```

```python
df2.head()
```

```python
df2 = df2.fillna(0)
```

```python
df2['Light Commercial']=df2['Light Commercial'].round(0).astype(int)+df2['LCV**'].round(0).astype(int)+df2['LCV'].round(0).astype(int)+df2["LCV *"].round(0).astype(int)
df2['Heavy Commercial']=df2['Heavy Commercial'].round(0).astype(int)+df2['Buses and coaches'].round(0).astype(int)+df2['Trucks (light up to heavy)***'].round(0).astype(int)+df2["Buses and coaches****"].round(0).astype(int)+df2["HCV"].round(0).astype(int)+df2["Med. HCV/Buses and coaches"].round(0).astype(int)
df2['Passenger cars'] = df2['Passenger cars'].round(0).astype(int)+df2['Passenger cars*'].round(0).astype(int)
```

```python
df2['SUV']=df2['SUV'].round(0).astype(int)
df2['Total']=df2['Total'].round(0).astype(int)
df2['Year']=df2['Year'].round(0).astype(int)
```

```python
Num_cars =df2.iloc[0:,0:7]
```

Change the country name to alpha2

```python
import pycountry
```

```python
def country_name(country_name):
    try:
        return pycountry.countries.get(name=country_name).alpha_2
    except:
        return ("not founded!")
```

```python
Num_cars['Country_alpha2'] = Num_cars.apply(lambda row: country_name(row.Country) , axis = 1)
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)#cite_note-16'
df = pd.read_html(url)
GDP = df[2]
```

```python
GDP = GDP.iloc[0:,[0,1,6,7]]
```

```python
GDP.head()
```

```python
GDP.columns
```

```python
GDP.columns = ['Country', 'UN Region', 'Estimate', 'Year']
GDP.columns
```

```python
import re
def remove_annotations(text):
    pattern = r'\[n \d+\]'
    return re.sub(pattern, '', text)
```

```python
GDP['Estimate'] = GDP['Estimate'].apply(remove_annotations)
GDP['Year'] = GDP['Year'].apply(remove_annotations)
```

```python
GDP['Country_alpha2'] = GDP.apply(lambda row: country_name(row.Country) , axis = 1)
```
